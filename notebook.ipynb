```python
import matplotlib.pyplot as plt

miesiace = ['Sty', 'Lut', 'Mar', 'Kwi', 'Maj', 'Cze',
            'Lip', 'Sie', 'Wrz', 'Paź', 'Lis', 'Gru']
temperatura = [-2, 0, 5, 12, 18, 22, 25, 24, 18, 11, 5, 0]

plt.figure(figsize=(10, 5))
plt.plot(miesiace, temperatura, marker='o', color='tomato', linewidth=2)
plt.fill_between(range(len(miesiace)), temperatura, alpha=0.2, color='tomato')
plt.title('Średnia temperatura w Warszawie')
plt.ylabel('Temperatura (°C)')
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```
